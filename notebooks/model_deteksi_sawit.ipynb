```python
import os
import shutil

import matplotlib as plt
import numpy as np
import tensorflow as tf
```

```python
train_dir = '../datasets/uav_palm_data/train2017'
val_dir = '../datasets/uav_palm_data/val2017'

dataset_dir = '../datasets/uav_palm_data/dataset'
os.makedirs(dataset_dir, exist_ok=True)
```

```python
for category in os.listdir(train_dir):
    category_dir = os.path.join(train_dir, category)
    if os.path.isdir(category_dir):
        shutil.copytree(category_dir, os.path.join(dataset_dir, category), dirs_exist_ok=True)

for category in os.listdir(val_dir):
    category_dir = os.path.join(val_dir, category)
    if os.path.isdir(category_dir):
        shutil.copytree(category_dir, os.path.join(dataset_dir, category), dirs_exist_ok=True)
```
